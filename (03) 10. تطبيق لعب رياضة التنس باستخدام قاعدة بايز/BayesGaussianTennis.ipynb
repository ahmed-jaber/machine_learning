```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('play_tennis_train_numeric.csv')

le_outlook = LabelEncoder()
le_outlook.fit(df['outlook'])

le_wind = LabelEncoder()
le_wind.fit(df['wind'])

le_play = LabelEncoder()
le_play.fit(df['play'])

df['outlook_num'] = le_outlook.transform(df['outlook']) 
df['wind_num'] = le_wind.transform(df['wind'])
df['play_num'] = le_play.transform(df['play'])
df
```

```python
from sklearn.naive_bayes import CategoricalNB,GaussianNB

X_cat = df[['outlook_num','wind_num']]

X_num = df[['temp','humidity']]

y = df['play_num']


model_cat = CategoricalNB(alpha=0)

model_num = GaussianNB()

model_cat.fit(X_cat, y)

model_num.fit(X_num, y)
```

```python
newDay = ['Sunny', 66, 90, 'Strong']

outlook_num = le_outlook.transform([newDay[0]])[0]

wind_num = le_wind.transform([newDay[3]])[0]
 
newDay_num = [[outlook_num, wind_num]]

new_Day_num_df = pd.DataFrame(newDay_num, columns=['outlook_num','wind_num'])

predclass_num = model_cat.predict(new_Day_num_df)

predclass = le_play.inverse_transform(predclass_num)
print(predclass)

pro_cat = model_cat.predict_proba(new_Day_num_df)
print(pro_cat)
```

```python
tem_num_cont = newDay[1]

humidity_num_cont = newDay[2]

newDay_num_cont = [[tem_num_cont,humidity_num_cont]]

new_Day_num_df = pd.DataFrame(newDay_num_cont, columns=['temp','humidity'])

pred_num = model_num.predict(new_Day_num_df)
pred_class = le_play.inverse_transform(pred_num)
print(pred_class)

pro_num = model_num.predict_proba(new_Day_num_df)
print(pro_num)
```

```python
pno_cat = pro_cat[0][0]

pyes_cat = pro_cat[0][1]

pno_num = pro_num[0][0]

pyes_num = pro_num[0][1]

pno = pno_num * pno_cat
pyes = pyes_num * pyes_cat

pno = pno / (pno + pyes)
pyes = pyes / (pno + pyes)

if (pno > pyes):
    print ("No", round(pno,4))
else:
    print ("Yes", round(pyes,4))
```
