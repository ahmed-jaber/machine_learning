```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('play_tennis_train.csv')

le_outlook = LabelEncoder()
le_outlook.fit(df['outlook'])

le_temp = LabelEncoder()
le_temp.fit(df['temp'])

le_humidity = LabelEncoder()
le_humidity.fit(df['humidity'])

le_wind = LabelEncoder()
le_wind.fit(df['wind'])

le_play = LabelEncoder()
le_play.fit(df['play'])

df['outlook_num'] = le_outlook.transform(df['outlook'])
df['temp_num'] = le_temp.transform(df['temp'])
df['humidity_num'] = le_humidity.transform(df['humidity'])
df['wind_num'] = le_wind.transform(df['wind'])
df['play_num'] = le_play.transform(df['play'])
df
```

```python
from sklearn.naive_bayes import CategoricalNB

X = df[['outlook_num', 'temp_num', 'humidity_num', 'wind_num'] ]

y = df['play_num']

model = CategoricalNB(alpha=0)

model.fit(X , y)
```

```python
newDay = ['Sunny','Cool', 'High','Strong']

outlook_num = le_outlook.transform([newDay[0]])[0]

temp_num = le_temp.transform([newDay[1]])[0]

humidity_num = le_humidity.transform([newDay[2]])[0]

wind_num = le_wind.transform([newDay[3]])[0]

new_Day_num = [[outlook_num, temp_num, humidity_num, wind_num]]
print(new_Day_num)

new_Day_num_df = pd.DataFrame(new_Day_num, columns=['outlook_num', 'temp_num', 'humidity_num', 'wind_num'])

predclass_num = model.predict(new_Day_num_df)

print(predclass_num)

predclass = le_play.inverse_transform(predclass_num)
print(predclass)
```

```python
prob = model.predict_proba(new_Day_num_df)
print(prob)
```

```python
y_pred = model.predict(X)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('\nAccuracy: {:.2f}\n'.format(100*accuracy_score(y, y_pred)))

print('\nPrecision: {:.2f}\n'.format(100*precision_score(y, y_pred)))

print('\nRecall: {:.2f}\n'.format(100*recall_score(y, y_pred)))

print('\nF1: {:.2f}\n'.format(100*f1_score(y, y_pred)))
```
