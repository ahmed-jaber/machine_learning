```python
import matplotlib.pyplot as plt

X = [[5,3],
[10,15],
[15,12],
[24,10],
[30,45],
[85,70],
[71,80],
[60,78],
[55,52],
[80,91]]

X1=[i[0] for i in X]
X2=[i[1] for i in X]

plt.scatter(X1,X2)
plt.show()
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=2)

model.fit(X)

print('centers',model.cluster_centers_)
print('labels',model.labels_)
```

```python
colors=model.labels_
plt.scatter(X1,X2,c=colors)

CX1=model.cluster_centers_[:,0]
CX2=model.cluster_centers_[:,1]
plt.scatter(CX1,CX2,color='green')
print(type(model.cluster_centers_))
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg=silhouette_score(X,model.labels_)
print(silhouette_avg)
```

```python
model=KMeans(n_clusters=3)
model.fit(X)

plt.scatter([i[0] for i in X],[i[1] for i in X],c=model.labels_,cmap='rainbow')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='black')
```

```python
silhouette_avg=silhouette_score(X,model.labels_)
print(silhouette_avg)
```

```python
range_n_cluster=[2,3,4,5]
for n_cluster in range_n_cluster:
    model=KMeans(n_clusters=n_cluster)
    model.fit(X)
    silhouette_avg=silhouette_score(X,model.labels_)
    print('For n_clusters =',n_cluster,'The average silhouette_score is :', silhouette_avg)
```
