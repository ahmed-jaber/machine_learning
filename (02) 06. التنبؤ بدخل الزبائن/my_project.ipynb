```python
import pandas as pd

df=pd.read_csv('income.csv')

df.info()
```

```python
df.describe()
```

```python
from pandas.plotting import scatter_matrix

df=df[['Age','Education','Gender','Income']]

scatter_matrix(df,figsize=(8,8))
```

```python
X=df[['Age','Education']]
y=df['Income']

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)


model=LinearRegression()
model.fit(X_train,y_train)

print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
y_pred=model.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math

print ('MSE :', round(mean_squared_error (y_pred, y_test), 2))
print ('RMSE :', round(math.sqrt(mean_squared_error (y_pred, y_test)), 2))
print ('MAE :', round(mean_absolute_error(y_pred, y_test), 2))
```
