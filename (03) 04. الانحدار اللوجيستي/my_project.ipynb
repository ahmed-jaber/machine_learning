```python
import pandas as pd
df=pd.read_csv('income.csv')
X=df[['Age','Education']]
y=df['VIP']
```

```python
dfVIP=df[df['VIP']==1]
dfNotVIP=df[df['VIP']==0]

import matplotlib.pyplot as plt

plt.scatter(dfVIP['Education'],dfVIP['Age'],color='red',marker='x')
plt.scatter(dfNotVIP['Education'],dfNotVIP['Age'],color='blue',marker='x')
plt.legend(['VIP','NotVIP'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X,y)

b0=model.intercept_
b1=model.coef_[0]
b2=model.coef_[1]

print('intercept=',b0)
print('coef b1= ',b1)
print('coef b2= ',b2)
```

```python
import math

def sigmoid(z):
    return 1/(1 + pow(math.e,-z))

Age=60
Education=12
Boundary=0.5

y_lr=b0+b1*Age+b2*Education

sig=sigmoid(y_lr)

if sig>Boundary:
    print('VIP')
else:
    print('NotVIP')
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,y)
```

```python
Age=60
Education=12

ex=[[Age,Education]]

income_info=pd.DataFrame(ex,columns=['Age','Education'])

pred=model.predict(income_info)

if pred==1:
    print('VIP')
else:
    print('NotVIP')

proba=model.predict_proba(income_info)
proba
```
