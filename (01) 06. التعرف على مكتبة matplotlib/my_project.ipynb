```python
import pandas as pd

df=pd.read_csv('failures.csv')
df
```

```python
import matplotlib.pyplot as plt

xPoints=df['Age']
yPoints=df['Failures']

plt.plot(xPoints,yPoints,marker='o',color='red')
plt.xlabel('Age')
plt.ylabel('Failures')
```

```python
x=df['Machine']
y=df['Failures']

plt.xlabel('Machine')
plt.xlabel('Failures')
plt.bar(x,y)
plt.show()
```

```python
x=df['Machine']
y=df['Failures']

plt.pie(y,labels=x,autopct='%1.1f')
plt.show()
```

```python
x=df['Age']
y1=df['Failures']
y2=df['DaysOff']

plt.subplot(1,2,1)
plt.scatter(x,y1)
plt.xlabel('Age')
plt.ylabel('Failures')

plt.subplot(1,2,2)
plt.scatter(x,y2)
plt.xlabel('Age')
plt.ylabel('DaysOff')

plt.show()
```

```python
df=pd.read_csv('stdMarks.csv')
dfCorr=df[['Math','Phy','Lang']]
correlations=dfCorr.corr(method='pearson')
correlations
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
cax=ax.matshow(correlations,vmin=-1,vmax=1)
fig.colorbar(cax)

ticks=[0,1,2]

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(['Math','Phy','Lang'])
ax.set_yticklabels(['Math','Phy','Lang'])

plt.show()
```
