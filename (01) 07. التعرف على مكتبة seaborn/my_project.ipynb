```python
import pandas as pd

df=pd.read_csv('titanic.csv')
df.head()
```

```python
import seaborn as sns

sns.countplot(x='Sex',data=df)
```

```python
sns.countplot(x='Sex',data=df,hue='Survived')
```

```python
sns.displot(x='Age',data=df,color='blue',bins=10)
```

```python
sns.displot(x='Age',data=df,color='blue',bins=10,hue=df['Survived'],kde=True)
```

```python
sns.lineplot(x=df['Age'],y=df['Weight'])
```

```python
sns.lineplot(x=df['Age'],y=df['Weight'],hue=df['Sex'])
```

```python
sns.scatterplot(x=df['Age'],y=df['Weight'])
```

```python
sns.scatterplot(x=df['Age'],y=df['Weight'],hue=df['Sex'])
```

```python
sns.boxplot(data=df,y='Age')
```

```python
sns.boxplot(x=df['Sex'],y=df['Age'])
```

```python
sns.barplot(data=df,x='Pclass',y='Age')
```
